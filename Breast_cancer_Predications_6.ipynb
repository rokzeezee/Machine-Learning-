```python
#split data into Independent (x) and dependent (y)
x = df[df.columns[:-1]].values
y = df["diagnosis"].values 
x.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
# 60% for tranning and 40% for testing 
```

```python
x_train, x_test, y_train, y_test  = train_test_split(x, y , test_size= 0.25 , random_state= 0)

 

```

```python
# scaling the data (feature scaling )
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_trian = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#creating models
#create a function 
```

```python
from pandas.core.common import random_state
def model (x_train ,y_train):
  #logistic regression
  from sklearn.linear_model import LogisticRegression
  log=LogisticRegression(random_state=0)
  log.fit(x_train,y_train)
  #desicion tree 
  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier(criterion = "entropy", random_state=0)
  tree.fit(x_train , y_train)
  #random forest classifier 
  from sklearn.ensemble import RandomForestClassifier
  forest = RandomForestClassifier(n_estimators=1 ,criterion="entropy" , random_state = 0)
  forest.fit(x_train , y_train )

  #print the model accuracy on the tranning data

  print("[0]LogisticRegression  Accuracy score  :" ,log.score , (x_train ,y_train) )
  print("[1]DecisionTreeClassifier  Accuracy score  :", tree.score , (x_train ,y_train) )
  print("[2]RandomForestClassifier  Accuracy score  :", forest.score , (x_train ,y_train) )
  
  return log ,tree,forest 
```

```python
#getting all models 
models= model(x_train ,y_train)
```

```python
#test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix( y_test ,models[0] .predict(x_test))

TP = cm[0][0]
TN = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]


print(cm)
print("TESTING ACCURACY = " , (TP + TN) / (TP +TN + FN + FP))
```

```python
 for i in range (len(models)):
   print("models")
   cm = confusion_matrix( y_test ,models[i] .predict(x_test))

   TP = cm[0][0]
   TN = cm[1][1]
   FN = cm[1][0]
   FP = cm[0][1]


   print(cm)
   print("TESTING ACCURACY = " , (TP + TN) / (TP +TN + FN + FP))
   print()
```
